Q3. Which event tends to attract more students from specific fields of study?



```python
event_field_count = df.groupby(['Events', 'Designation']).size().reset_index(name='Student Count')
event_field_count
```

```python
top_events = event_field_count.sort_values(by='Student Count', ascending=False)
top_events
```

```python
plt.figure(figsize=(12, 8))
for designation in event_field_count['Designation'].unique():
    subset = event_field_count[event_field_count['Designation'] == designation]
    plt.bar(subset['Events'], subset['Student Count'], label=designation)

plt.xlabel("Event")
plt.ylabel("Number of Students")
plt.title("Event-wise Student Participation by Designation")
plt.xticks(rotation=90)
plt.legend(title='Designation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Q8. Do students with high CGPA and more experience in programming languages have higher expected salaries? (Find the average)

```python
high_cgpa = df['CGPA'] >= 8
high_experience = df['Experience with python (Months)'] >= 6
```

```python
filtered_students = df[high_cgpa & high_experience]
filtered_students.head()
```

```python
average_salary = filtered_students['Expected salary (Lac)'].mean()
average_salary
```

```python
overall_avg_salary = df['Expected salary (Lac)'].mean()
overall_avg_salary
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(['High CGPA & Experience', 'Overall Average'],
        [average_salary, overall_avg_salary])
plt.ylabel("Expected Salary")
plt.title("Salary Comparison Based on CGPA & Experience")
plt.show()
```
